```python
#构建词袋模型
dictionary = corpora.Dictionary(corpus)
print(dictionary)
```

```python
doc_vectors = [dictionary.doc2bow(text) for text in corpus]
print(len(doc_vectors))
print (doc_vectors)
```

```python
#建立TF-IDF模型
tfidf = models.TfidfModel(doc_vectors)
print(tfidf)
tfidf_vectors = tfidf[doc_vectors]
for doc in tfidf_vectors:
    print(doc)
```

```python
print (len(tfidf_vectors))
print (len(tfidf_vectors[2]))
```

```python
#构建一个query文本，是高血压主题的，利用词袋模型的字典将其映射到向量空间
query = tokenization('/Users/xieli/Desktop/policy/data/国务院应对新型冠状病毒感染肺炎疫情联防联控机制关于印发企事业单位复工复产疫情防控措施指南的通知.txt')
```

```python
query_bow = dictionary.doc2bow(query)
```

```python
query_model = tfidf[query_bow]
print(query_model)
```

```python
index = similarities.MatrixSimilarity(tfidf_vectors)
```

```python
#用TF-IDF模型,使用余弦相似度来度量两个向量之间的相似.
sims = index[query_model]
print(sims)
print(list(enumerate(sims)))
```
